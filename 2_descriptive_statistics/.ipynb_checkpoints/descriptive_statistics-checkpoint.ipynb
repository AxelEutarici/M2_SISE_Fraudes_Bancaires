```python
data = df.select_dtypes(include=['float64', 'int64'])
```

```python
desc = data.describe()
print(desc)
# Calculer le mode
mode = data.mode()
print(mode)
# Calculer la variance
var = data.var()
print(var)
```

```python
# Calculer la corrÃ©lation entre les variables explicatives quantitatives
corr = data.corr()
print(corr)
sns.heatmap(corr, cmap='coolwarm')
```

```python
skewness = data.skew()
kurtosis = data.kurtosis()

#Calculer les valeurs atypiques (outliers)
z_scores = stats.zscore(data)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
data_no_outliers = data[-filtered_entries]
data_no_outliers
```

```python
##### target = df['FlagImpaye']
corr = data.corrwith(target)
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.bar(corr.index, corr.values)
plt.xticks(range(len(corr)), corr.index, rotation=45, fontsize=8)
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficients')
plt.savefig("corrwith_plot.png")
plt.show()
print(corr)
```
